```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/no_live_data.csv",encoding="ISO-8859-1",
                         dtype={'CustomerID': str,'InvoiceID': str})

df.insert(0, "id", df.index + 1)
df.head()
```

```python
invoice = df[['id', 'InvoiceNo', 'StockCode','Description','Quantity','InvoiceDate']]
invoice.head()
```

```python
customer = df[['id','UnitPrice','CustomerID','Country']]
customer.head()
```

```python
df = df.drop(['id'], axis=1)
df.head()
```

```python
result = pd.merge(left=invoice, right=customer, left_on='id', right_on='id', how='inner')
result = result.drop(['id'], axis=1)
result.head()
```

```python
print(df.shape)
print(customer.shape)
print(invoice.shape)
print(result.shape)
```

```python
df.equals(result)
```

```python
invoice.to_csv('../data/invoice_data.csv', index=False)
customer.to_csv('../data/customer_data.csv', index=False)
```
